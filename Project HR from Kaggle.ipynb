Here EmployeeCount is having a number 1 for all the rows, EmployeeNumber is also having less valuable information, StandardHours also have same number 80 for all rows. Where mean median different percentile values are same mostly all the data of those columns are same. So for training and testing model we can remove these unnecessary data fields.

```python
#let us remove three unnecessary data fileds
drop_list = ["EmployeeCount", "EmployeeNumber", "StandardHours"]
columns_all = list(df.describe().columns)
#col_categorical = list(set(df.columns).difference(columns_all))
col_numerical = [e for e in columns_all if e not in drop_list]
```

```python
#Now we will only select the columns that has data type as object not int64
categorical_col = []
for k,v in df.iteritems():
    if v.dtype == 'object':
        categorical_col.append(k)
categorical_col
```

Now we will see the correlation between different columns to see which of this could be our target and which can be train and test data sets.

```python
df[col_numerical].corr() #this will print out the correlation between columns
```

```python
#now we will see the heatmap
pt.figure(figsize=(24,10))
sb.heatmap(df[col_numerical].corr(), annot=True, fmt=".2f")
```

From the above heatmap we can see these strong correlations (JobLevel,Age,MonthlyIncome,TotalWorkingYears) (JobLevel,MonthlyIncome, YearsAtCompany,YearsInCurrentRole,YearsSinceLastPromotion,YearsWithCurrManager)

```python
#we have to mask for the upper triangle
mask = np.zeros_like(df[col_numerical].corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
pt.figure(figsize=(24,10))
cmap = sb.diverging_palette(220, 10, as_cmap=True) #this will generate a custom color map
#cmap = sb.light_palette((210, 90, 60), input="husl") #this will generate a custom color map
sb.heatmap(df[col_numerical].corr(), mask=mask, cmap=cmap, vmax=.3, center=0, annot=True, fmt=".2f")
```
